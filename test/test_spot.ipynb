```python
import toml
import okx.Trade as Trade

# params
api_config = toml.load('../config/api.toml')
api_key = api_config['api_key']
secret_key = api_config['secret_key']
passphrase = api_config['passphrase']
demo_flag = '0'
domain = 'https://www.dojqutmbiuuo.com'
debug_flag = False
```

## spot market data

```python
import okx.MarketData as MarketData

market_api = MarketData.MarketAPI(flag=demo_flag, domain=domain, debug=debug_flag)

result = market_api.get_tickers(instType='SPOT')
print(result)
result = market_api.get_ticker(instId='SOL-USDT')
print(result)
```

## get funding data

```python
import okx.Funding as Funding

funding_api = Funding.FundingAPI(api_key, secret_key, passphrase,
                                 use_server_time=False, flag=demo_flag,
                                 domain=domain, debug=debug_flag)
result = funding_api.get_balances()
print(result)
```

## get balance

```python
import okx.Account as Account

account_api = Account.AccountAPI(api_key, secret_key, passphrase,
                                 use_server_time=False, flag=demo_flag,
                                 domain=domain, debug=debug_flag)
result = account_api.get_account_balance()
data = result['data'][0]  # sub-account 
details = data['details']  # sorted by diff ccy
```

# get trading pairs

```python
import okx.PublicData as PublicData

public_data_api = PublicData.PublicAPI(api_key, secret_key, passphrase,
                                       use_server_time=False, flag=demo_flag,
                                       domain=domain, debug=debug_flag)
result = public_data_api.get_instruments(instType='SPOT')
result
```
